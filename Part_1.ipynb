### 3. Task
Now it is time to split the data, into a training, evaluation and test set. As a reminder, the date we are trying to predict snow fall for is the following, and hence should constitute your test set.

```python
import datetime, timedelta
# +3 leap years, -1 - tomorrow 
wanted_date = str(datetime.datetime.today()- datetime.timedelta(days=11*365 + 3 - 1)).split(' ')[0]
print("Wanted date: " + wanted_date + ".")
```

```python
all_dates = weather_data_cleaned["date"].dt.date

min_date = str(all_dates.min())
max_date = str(all_dates.max())
print("The available dates are between " + min_date + " and " + max_date + ".")
print("So, the latest date is " + max_date + " and it is earlier as our wanted date " + wanted_date + ".")
```

This means that we do not have to pay attention to which data is saved in which set.

My idea is to make the snowfall forecast "overnight". So, we have all the data from day (n) and want to know whether it will snow on the (n + 1)-th day. For this I have to expand the data with the column "snow tomorrow", where it is saved whether it will snow the next day. Then I can randomly split the data into subsets.

```python
# Create new column "yesterday"
yesterday = pd.to_datetime(weather_data_cleaned.date) - datetime.timedelta(days=1)
weather_data_cleaned['yesterday'] = yesterday
```

```python
# Create new dataframes to save dates and stations with or without snowfall
day_station = weather_data_cleaned.filter(['yesterday', 'station_number', 'snow'])
day_station_snow = day_station.loc[day_station['snow'] == 1]
day_station_no_snow = day_station.loc[day_station['snow'] == 0]
```

```python
# Merge data: now we have the days, after then it snowed / not snowed
weather_data_cleaned_snow = weather_data_cleaned[weather_data_cleaned.set_index(['date','station_number']).index.isin(day_station_snow.set_index(['yesterday','station_number']).index)]
weather_data_cleaned_no_snow = weather_data_cleaned[weather_data_cleaned.set_index(['date','station_number']).index.isin(day_station_no_snow.set_index(['yesterday','station_number']).index)]
```

```python
weather_data_cleaned_snow['snow_tomorrow'] = 1
weather_data_cleaned_no_snow['snow_tomorrow'] = 0
```

```python
# Merge data
weather_data_cleaned_all = weather_data_cleaned_snow.append(weather_data_cleaned_no_snow, ignore_index=True)
```

```python
weather_data_cleaned_all = weather_data_cleaned_all.drop([
    'yesterday',
], axis = 1)
```

As we can see, there are significantly fewer days with snowfall than without. It is important that this ratio is preserved after dividing into subsets. 

```python
def split_train_eval_test(df, frac_train=0.65, frac_eval=0.1, frac_test=0.25):
    '''
    Splits a Pandas dataframe into training, evaluation and test sets.
    Each subset size can be set.
    Each set has the same relative frequency of the values in the column 'snow_tomorrow'.
    '''
    if frac_train + frac_eval + frac_test != 1.0:
        raise ValueError('fractions %f, %f, %f do not add up to 1.0' % \
                         (frac_train, frac_eval, frac_test))

    stratify_col = 'snow_tomorrow'
    
    X = df # All columns
    #X = df.drop([stratify_col], axis=1) # All columns, except 'snow_tomorrow'
    y = df[[stratify_col]] # Dataframe of just the 'snow_tomorrow' column

    # Split input dataframe into train and eval_test dataframes.
    train, eval_test, y_train, y_eval_test = train_test_split(X,
                                                          y,
                                                          stratify=y,
                                                          test_size=(1.0 - frac_train))

    # Split the eval_test dataframe into eval and test dataframes.
    relative_frac_test = frac_test / (frac_eval + frac_test)
    _eval, test, y_eval, y_test = train_test_split(eval_test,
                                                      y_eval_test,
                                                      stratify=y_eval_test,
                                                      test_size=relative_frac_test)
    
    X_train = train.drop([stratify_col], axis=1) # All columns, except 'snow_tomorrow'
    X_eval = _eval.drop([stratify_col], axis=1) # All columns, except 'snow_tomorrow'
    X_test = test.drop([stratify_col], axis=1) # All columns, except 'snow_tomorrow'
    
    assert len(df) == len(X_train) + len(X_eval) + len(X_test)

    return train, _eval, test
```

```python
train, _eval, test = split_train_eval_test(weather_data_cleaned_all)
```

```python
# Save data as .csv to use it in part 2

weather_data_cleaned_all.to_csv('weather_data_cleaned_all.csv')
train.to_csv('train.csv')
_eval.to_csv('_eval.csv')
test.to_csv('test.csv')
```
